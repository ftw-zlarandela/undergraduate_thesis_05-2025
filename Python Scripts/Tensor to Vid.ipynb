```python
import torch
import numpy as np
import cv2
```

```python
def tensor_to_video(hsv_tensor, output_path="output.mp4", fps=60):

    num_frames, _, height, width = hsv_tensor.shape
    fourcc = cv2.VideoWriter_fourcc(*'x264')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    for frame_idx in range(num_frames):
        hsv_frame = hsv_tensor[frame_idx].cpu().numpy().astype(np.uint8)  # Convert tensor to NumPy
        hsv_frame = np.transpose(hsv_frame, (1, 2, 0))
        # Convert HSV back to BGR for saving
        bgr_frame = cv2.cvtColor(hsv_frame, cv2.COLOR_HSV2BGR)

        out.write(bgr_frame)

    out.release()
```

```python
names = ["aspen", "crowd2", "ducks", "life1", "rush1", "pedestrian2", "sintel3", "sunflower1"]
```

```python
for name in names:
    hsv_tensor = torch.load(fr"D:\Users\Acer\Desktop\Thesis\{name}_hsv_tensor", weights_only=True)
    tensor_to_video(hsv_tensor, f'{name}_0418.mp4')
    del hsv_tensor

    print(f'{name} done')
```
