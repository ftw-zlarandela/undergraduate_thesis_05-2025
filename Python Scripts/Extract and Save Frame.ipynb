```python
import numpy as np
import cv2
```

```python
#names = ["sunflower1"]
names = ["crowd2", "rush1"]
#attacks = ["raw", "watermarked"]
attacks = ["raw", "45", "90", "135", "swapping", "static", "vhs", "vhs2"]
```

```python
for name in names:
    for attack in attacks:
        video_path = fr"D:\Users\Acer\Desktop\Thesis_0419\recovered_watermark\{name}_{attack}_rec_wm.mp4"
        cap = cv2.VideoCapture(video_path)
        cap.set(cv2.CAP_PROP_POS_MSEC, 0)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite(f"{name}_{attack}_watermark_0.png", frame)
        print(f'{name}_{attack} done')
    print(f'{name} all done')
```
