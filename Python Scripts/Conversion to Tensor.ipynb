```python
import torch
import numpy as np
import cv2
```

```python
def conv_vid_hsv(vid_path, channel1, channel2):
    cap = cv2.VideoCapture(vid_path)

    if not cap.isOpened():
        print(f"Error: Could not open video file {vid_path}")
        return None

    frames = []
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        # Convert default channels BGR to Hue, Saturation, Value
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        # Convert to tensor
        frame = torch.tensor(frame, dtype=torch.torch.float) # / 255
        # Convert (Height, Width, Channel) into (Channel, Height, Width)
        frame = frame.permute(2, 0, 1)[channel1:channel2+1]
        #print(frame.shape)
        frame = torch.reshape(frame, (channel2-channel1+1, frame.shape[-2], frame.shape[-1]))
        #print(frame.shape)
        frames.append(frame)

    cap.release()

    frames = frames[:300]
    frames = torch.stack(frames)
    print(frames.shape)

    return frames
```
