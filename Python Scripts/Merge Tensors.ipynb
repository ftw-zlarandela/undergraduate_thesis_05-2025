```python
import torch
import numpy as np
```

```python
def merge_hsv_to_tensor(hs_tensor, v_tensor):
    hsv_tensor = torch.cat([hs_tensor, v_tensor], dim=1)
    return hsv_tensor
```

```python
names = [  "pedestrian2", "sintel3", "sunflower1"]
names = ["aspen", "crowd2", "ducks", "life1", "pedestrian2", "rush1", "sintel3", "sunflower1"]
names = ["aspen"]
```

```python
for name in names:
  embedded_tensor = torch.load(f'/content/drive/Shareddrives/thesis_0419 merge_tensors/{name}_embedded', map_location=torch.device('cpu'), weights_only=True)*255.0
  in_hs = torch.load(f'/content/drive/MyDrive/thesis/testing_hs_tensors/{name}_hs_tensor', weights_only=True)

  hsv_tensor = merge_hsv_to_tensor(in_hs, embedded_tensor)
  del in_hs, embedded_tensor

  torch.save(hsv_tensor.detach().clone(), f'{name}_hsv_tensor')
  del hsv_tensor

  print(f'{name} done')
```
