```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Metrics:

```python
def get_frames(video_path, idx):
  cap = cv2.VideoCapture(video_path)
  cap.set(cv2.CAP_PROP_POS_MSEC, idx-1)
  ret, frame = cap.read()
  return frame
```

```python
def nc(watermark_vid, recovered_vid):
    i = 0
    nc_i = 0

    while i != 300:
      nc_i += cv2.matchTemplate(get_frames(watermark_vid, i), get_frames(recovered_vid, i), cv2.TM_CCORR_NORMED)
      i += 1

    nc = nc_i/300
    return nc
```

```python
def ber(watermark_vid, recovered_vid):
  # Evaluation for original watermark and recovered watermark after attacks
    i = 0
    ber_i = 0

    while i != 300:
      watermark_bits = np.unpackbits(get_frames(watermark_vid, i).astype(np.uint8))
      recovered_bits = np.unpackbits(get_frames(recovered_vid, i).astype(np.uint8))

      # Compute BER
      bit_errors = np.sum(watermark_bits != recovered_bits)
      total_bits = len(watermark_bits)

      ber_i += bit_errors / total_bits
      i += 1

    ber = ber_i/300
    return ber
```
