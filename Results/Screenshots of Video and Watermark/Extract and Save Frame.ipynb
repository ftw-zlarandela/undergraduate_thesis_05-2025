```python
import numpy as np
import cv2
```

```python
names = ["crowd2", "rush1", "sunflower1"]
attacks = ["raw", "watermarked"]
#attacks = ["45", "90", "135", "swapping", "static", "vhs", "vhs2"]
```

```python
for name in names:
    for attack in attacks:
        video_path = fr"D:\Users\Acer\Desktop\Thesis_0419\testing videos\{name}_{attack}.mp4"
        cap = cv2.VideoCapture(video_path)
        cap.set(cv2.CAP_PROP_POS_MSEC, 0)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite(f'{name}_{attack}_0', frame)
        print(f'{name}_{attack} done')
    print(f'{name} all done')
```
